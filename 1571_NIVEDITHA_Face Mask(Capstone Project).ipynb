 #### Import necessary libraries

```python
import glob
```

```python
with_mask = glob.glob('with_mask 1/with_mask 1/*.jpg')
without_mask = glob.glob('without_mask/without_mask/*.jpg')
```

#### Store all the images inside a list and create a list with the corresponding labels

#### For Training set

```python
image = []
label = []

for i in with_mask:
    img = plt.imread(i)
    image.append(img)
    label.append(0)
```

```python
for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(image[i])
    plt.title(label[i])
    plt.tight_layout()
```

```python
for i in without_mask:
    img = plt.imread(i)
    image.append(img)
    label.append(1)
```

```python
label
```

#### For Testing set

#### Shuffle both image list and label list (In both Training and testing set)

##### Use shuffle function from sklearn utils

```python
from sklearn.utils import shuffle
```

```python
image, label = shuffle(image, label, random_state=100)
```

```python
len(image)
```

#### Plot a few images from our shuffled list with its corresponding label (Either from training or test set)

```python
for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(image[i])
    if label[i]==0:
        plt.title('with_mask')
    else:
        plt.title('without_mask')
    plt.tight_layout()
```

#### Check the shape of the first 10 images in our list (Either from training or test set)

```python
for i in range(10):
    print(image[i].shape)
```

#### Reshape the images ( For both training and testing set)

Also reassign them to new variables named X_train and X_test and convert them to a numpy array

```python
from skimage.transform import resize
```

```python
X =[]
for i in image:
    X.append(resize(i,(64,64,3)))
X = np.array(X)
```

#### create numpy arrays named y_train and y_test from our training and testing label lists we created earlier

```python
y = np.array(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,random_state=42,test_size=0.1)
```

#### Plot some of our reshaped images

```python
for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(xtrain[i])
    plt.title(ytrain[i])
    plt.tight_layout()
```

#### Check the shape of x_train and x_test and y_train and y_test

```python
print(f'Shape of xtrain is:',xtrain.shape)
print(f'Shape of xtest is:', xtest.shape)
print(f'Shape of ytrain is:',ytrain.shape)
print(f'Shape of ytest is:',ytest.shape)
```

#### Create a keras Sequential model with convolution layers, Flatten and Hidden layers

```python
from tensorflow import keras
```

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(16,(3, 3), activation = 'relu', input_shape = (64,64,3)),
    keras.layers.MaxPool2D(2,2),

    keras.layers.Conv2D(32,(3, 3), activation = 'relu'),
    keras.layers.MaxPool2D(2,2),

    keras.layers.Flatten(),
    keras.layers.Dense(64, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])
```

#### Compile the model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

#### Check the summary of our model

```python
model.summary()
```

#### Train the model.

```python
model.fit(xtrain, ytrain, epochs=15, validation_split=0.1)
```

#### Evaluate the model

```python
model.evaluate(xtrain, ytrain)
```

#### Make predictions with X_test

```python
xtest
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

#### Use np.argmax to fetch indices of max element of the array

Store the values in a list

#### Plot X_test with label from our prediction

```python
for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(xtest[i])
    plt.title(ypred[i])
    plt.tight_layout()
```
